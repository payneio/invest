# Fidelity — export (no API)

Fidelity has no programmatic fetch; you export by hand and **datestamp** both files
into `data/raw/fidelity/` (newest by name wins). Then open **`refresh.ipynb`** to
rebuild + reconcile.

1. **Positions CSV** — Accounts → Portfolio → Positions → Download →
   `data/raw/fidelity/YYYYMMDD_fidelity_positions.csv`
2. **Activity JSON** — load the unpacked extension in `fidelity-export-extension/`
   (`chrome://extensions` → Developer mode → Load unpacked), open the Activity page,
   run it, save `~/Downloads/fidelity_history.json`, then:
   ```
   mv ~/Downloads/fidelity_history.json data/raw/fidelity/YYYYMMDD_fidelity_history.json
   ```
   See [`fidelity-export-extension/README.md`](../fidelity-export-extension/README.md).
